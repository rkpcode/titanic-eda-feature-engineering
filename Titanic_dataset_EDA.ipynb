```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# loading dataset form seaborn
df = sns.load_dataset('titanic')


# type casting for memory optimization

df['sex'] = df['sex'].astype ('category')
df['survived'] = df['survived'].astype('int16')
df['pclass'] = df['pclass'].astype('int16')
df['age'] = df['age'].astype('float32')
df['sibsp'] = df['sibsp'].astype('int16')
df['parch'] = df['parch'].astype('int16')
df['fare'] = df['fare'].astype('float32')
df['embarked'] = df['embarked'].astype('category')
df['who'] = df['who'].astype('category')
df['embark_town'] = df['embark_town'].astype('category')
df['alive'] = df['alive'].astype('category')

#checking null value
#print(df.isnull().sum())

#visualizing null value
sns.heatmap(df.isnull(), cbar=False)
plt.title("Null Value Map")
plt.show()

#cleaning data
df['age'] = df['age'].fillna(df['age']. median())
df['embarked'] = df['embarked'].fillna(df['embarked'].mode()[0])
df['embark_town'] = df['embark_town'].fillna(df['embark_town'].mode()[0])
#lot of null value present in 'deck' we have to drop the column
df. drop(columns= 'deck',inplace =True)

#save clean data set

df.to_csv('clean_data.csv', index=False)

print (df.dtypes)
print (df.head())
```

```python
# Bi-variaite analysis(category v/s category)

#cat v/s cat analysis: who survived more male v/s female?
sns.countplot(x='sex', hue='survived', data=df)
plt.title("Gender vs Survival")
plt.show()

print ("\n","Female survival count is more.")

#pclass v/s survival
print ('\n',df.groupby('survived')['pclass'].value_counts())
sns.countplot(x= 'survived', hue='pclass', data=df)
plt.title("pclass vs Survival")
plt.show()

#embarked v/s survival
print ('\n',df.groupby('survived')['embarked'].value_counts())
sns.countplot(x= 'survived', hue='embarked', data=df)
plt.title("Embarked vs Survival")
plt.show()

#who v/s survival
print ('\n',df.groupby('survived')['who'].value_counts())
sns.countplot(x= 'survived', hue='who', data=df)
plt.title("Who vs Survival")
plt.show()
```

```python
#group by() Bi - variate Analysis

#gender wise survival rate
print (df.groupby('sex')['survived'].mean())

#Embarked port wise survival rate
print (df.groupby('embarked')['survived'].mean())


#pclass wise fare

print (df.groupby('pclass')['fare'].mean())

#Multiple grouping columns analysis

#1:gender+pclass wise survival rate

print (df.groupby(['sex','pclass'])['survived'].mean())

#2:Avg age by pclass and sex
print ('Avg age by pclass and sex','\n',df.groupby(['sex','pclass'])['age'].mean())


#Multiple summary stats by aggregation

print (df.groupby('pclass')['fare'].agg(['mean', 'max', 'min']))


#sorting group result
print ("Avarage Fare Class wise","\n",df.groupby('pclass')['fare'].mean().sort_values(ascending=False))

print ("Maximum Fare Class wise","\n",df.groupby('pclass')['fare'].max().sort_values(ascending=False))

print ('who gave maximum fare','\n',df.groupby('who')['fare'].max().sort_values(ascending=False))
```

```python

# group by () +Agg()+ lambda function

# Gender wise: average age, total fare paid, and survival %
df.groupby('sex').agg({
    'age': 'mean',
    'fare': 'sum',
    'survived': lambda x: round((x.sum() / x.count()) * 100, 2)
})
```

```python

# cross tab: compare two category col by creating frequency table

print ('How many males and females survived vs not survived ?','\n',pd.crosstab(index=df['sex'], columns=df['survived']))
sns.countplot(x='sex', hue='survived', data=df)
plt.title("Gender vs Survival")
plt.show()
```

```python

#pivot table:
#sex survival rate
print (pd.pivot_table(df, index='sex', values='survived', aggfunc='mean'))

#sex+pclass survival rate
print (pd.pivot_table(df,index='sex', columns='pclass', values= 'survived',aggfunc = 'mean'))
```

```python
# cat V/s num: age v/s Survival
sns.boxplot(x='survived', y = 'age', data = df)
plt.title("Age V/s Survival")
plt.show()
print ("Survival ke beech age mein koi bada difference nahi hee, lekin survivors mein thodi si young population ho sakti hai.")

sns.violinplot(x='survived', y = 'age', data = df)
plt.title("Age V/s Survival")
plt.show()

# cat v/s num analysis: Did Rich people survived ?
sns.boxplot(x='survived', y='fare', data = df)
plt.title("Fare vs Survival")
plt.show()

print ("\n","clearly yes, because box plot median line  and spread of '1' biger than that of '0'")

#num v/s num : is there any relationship between age and fare?
sns.scatterplot(x='age', y='fare', hue='survived', data=df)
plt.title("Age vs Fare by Survival")
plt.show()
print ("No Strong Linear Trend: There’s no clear linear relationship between age, fare, and survival. However, higher fares might slightly correlate with survival, especially for younger passengers.Outliers: A few points with very high fares (above 300) are visible, mostly among survivors (orange).")

#correlation analysis
correlation = df.corr(numeric_only=True)
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
import os
print("" + os.linesep + "")

#Multi-varaites analysis

#1:groupby() multiple columns
# For each 'sex','embarked' and 'pclass' combination, find:
# - average age
# - average fare
# - number of survivors
# - survival rate

df_grouped_result = df.groupby(['embarked', 'pclass','sex']).agg({
    'age' : ['mean','min', 'max'],
    'fare': 'mean',
    'survived': ['sum', 'mean', 'count']
}).sort_values(by = [('survived','mean')], ascending=False)

plt.figure(figsize=(10, 6))
sns.heatmap(df_grouped_result, annot=True, fmt=".2f", cmap="YlOrBr")
plt.title("Survival Rate by Embarked, Sex, and Pclass")
plt.ylabel("Embarked")
plt.xlabel("Sex & Pclass")
plt.show()

#2:Crosstab to analyze survival by Sex and Pclass
print (pd.crosstab(index=df['pclass'], columns=df['sex'], values=df['survived'], aggfunc='mean'))
survival_crosstab = pd.crosstab(index=df['embarked'], columns=df['sex'], values=df['survived'], aggfunc='mean', margins=True)
print (survival_crosstab)
# Heatmap banaate hain
plt.figure(figsize=(6, 4))
sns.heatmap(survival_crosstab, annot=True, cmap="YlGnBu", fmt=".2f")
plt.title("Survival Rate by Embarkation Port and Gender")
plt.ylabel("Embarked")
plt.xlabel("Sex")
plt.show()

# pivot table
pivot = df.pivot_table(
    values='survived',
    index='embarked',
    columns=['sex', 'pclass'],
    aggfunc='mean'
)
print (pivot)

# heatmap visualization
plt.figure(figsize=(10, 6))
sns.heatmap(pivot, annot=True, fmt=".2f", cmap="YlOrBr")
plt.title("Survival Rate by Embarked, Sex, and Pclass")
plt.ylabel("Embarked")
plt.xlabel("Sex & Pclass")
plt.show()
```
